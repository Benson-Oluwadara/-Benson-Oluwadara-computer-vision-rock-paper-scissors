```python
import cv2
from keras.models import load_model
import numpy as np
import random
import time

count=0
options=['rock','paper','scissors']
op=""
def video():
    model = load_model('keras_model.h5',compile=False)
    cap = cv2.VideoCapture(0)
    data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

    while True: 
        ret, frame = cap.read() 
        resized_frame = cv2.resize(frame, (224, 224), interpolation = cv2.INTER_AREA)
        image_np = np.array(resized_frame)
        normalized_image = (image_np.astype(np.float32) / 127.0) - 1 # Normalize the image
        data[0] = normalized_image
        prediction = model.predict(data)
        cv2.imshow('frame', frame)
        # Press q to close the window
        #print(prediction)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            #print(prediction)
            max_element=np.max(prediction) 
            #print(max_element)
            arr=np.array(prediction)
            if(max_element==arr[0][0]):
                print("I choose rock")
                op="rock"
                #return op
            elif(max_element==arr[0][1]):
                print("I choose paper")
                op="paper"
                #return op
            elif(max_element==arr[0][2]):
                print("I choose scissors")
                op="scissors"
                #return op
            else:
                print("I choose nothing")
                op="nothing"
                #return op               
            break
                    
    # After the loop release the cap object
    cap.release()
    # Destroy all the windows
    cv2.destroyAllWindows()
    return op

def get_computer_choice():
    op=random.choice(options)
    print("Computer Chooses",op)
    return op

def CountDown(seconds):
    tim = int(time.time())
    timer = 1
    while(timer != 0):
        tim2 = int(time.time())
        timer = (tim2 - tim) - seconds

def get_winner(user_choice,computer_choice):
    if(user_choice==computer_choice):
        print(f"Both players selected {user_choice}. It's a tie!")
    elif (user_choice=='rock'):
        if(computer_choice=='paper'):
            print("Paper covers rock! You lose.")
            return False
        elif(computer_choice=='scissors'):
            print("Rock smashes scissors! You win!")
            #count+=1
            return True
    elif(user_choice=='paper'):
        if(computer_choice=='rock'):
            print("Paper covers rock! You win!")
            #count+=1
            return True
        elif(computer_choice=='scissors'):
            print("Scissors cuts paper! You lose.")
    elif(user_choice=='scissors'):
        if(computer_choice=='paper'):
            print("Scissors cuts paper! You win!")
            #count+=1
            return True
        elif(computer_choice=='rock'):
            print("Rock smashes scissors! You lose.")
            
def play():
    return get_winner(video(),get_computer_choice())
list1=[]
while(True):
    CountDown(5)
    list1.append(play())
    if(list1.count(True)==3):
        break
#op=video()  
#print("XXXXX---XXXXX:",op)
```
